```python
import pandas as pd
import sqlite3
import re

dbname = 'nishioka-nakajima_6-3_6-7-4-6.db'
conn = sqlite3.connect(dbname)
c = conn.cursor()
df_score = pd.read_sql("select * from score", conn)
```

```python
#サーバーの格納しなおし（順番おかしい）

s=df_score['Game'].values.tolist()
p=df_score['Score'].values.tolist()
firstServer="Nakashima"
secondServer="Nishioka"
total=0
l_pre=0
r_pre=0
for i in range(len(s)):
    m=re.findall(r'(\d)-(\d)',s[i])
    m_point=re.findall(r'(\d+)-(\d+)',p[i])
    #print(m_point)
    if(m):
        l=int(m[0][0])
        r=int(m[0][1])
        g=l+r
        
        if(l!=l_pre or r!=r_pre):
            total=total+1
        
        if(total%2==0):
            if(df_score['Server'][i]!=firstServer):
                print(l,r,total)
                print(df_score['Server'][i])
                df_score['Server'][i]=firstServer
        else:
            if(df_score['Server'][i]==firstServer):
                #print(l,r,total)
                #print(df_score['Server'][i])
                df_score['Server'][i]=secondServer        
        if(l==6 and r==6):
            #print(m_point)
            p1=int(m_point[0][0])
            p2=int(m_point[0][1])
            print(p1,p2,total)
            if(total%2==0):
                if((p1+p2+1)%4<2):#0 1 firstserver
                    if(df_score['Server'][i]!=firstServer):
                        df_score['Server'][i]=firstServer
                else:# 2 3 secondserver
                    if(df_score['Server'][i]==firstServer):
                        df_score['Server'][i]=secondServer
            else:
                if((p1+p2+1)%4>1):#2 3⇒firstserver 
                    if(df_score['Server'][i]!=firstServer):
                        df_score['Server'][i]=firstServer
                else:#0 1 secondserver
                    if(df_score['Server'][i]==firstServer):
                        df_score['Server'][i]=secondServer           
            
        l_pre=l
        r_pre=r
        
df_score.to_csv("fix_score.csv")
            
```
